```python
from skimage.io import imread, imsave, imshow
```

```python
img = imread('tiger-color.png')
```

```python
img.shape
```

```python
from skimage import img_as_float, img_as_ubyte
```

```python
img_f = img_as_float(img)
```

```python
import numpy as np
```

```python
#r, g, b = np.dsplit(img_f, 3)
```

```python
img_y = 0.2126 * img_f[:, :, 0] + 0.7152 * img_f[:, :, 1] + 0.0722 * img_f[:, :, 2]
```

```python
img_y.dtype
```

```python
imshow(img_y);
```

```python
img_y.shape
```

```python
img_y_byte = img_as_ubyte(img_y)
```

```python
img_y_byte.dtype
```

```python
imshow(img_y_byte);
```

```python
imsave('tiger-y-my.png', img_y_byte)
```

```python
img_y_byte.shape
```

```python
tiger_y = imread('tiger-y.png')
```

```python
np.array_equal(img_y_byte, tiger_y)
```
