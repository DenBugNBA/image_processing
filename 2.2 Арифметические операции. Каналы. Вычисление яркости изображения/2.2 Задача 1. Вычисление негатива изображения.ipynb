```python
from skimage.io import imread, imsave, imshow
```

```python
img = imread('tiger-color.png')
```

```python
img_negative = imread('tiger-negative.png')
```

```python
imshow(img_negative);
```

```python
from skimage import img_as_float
```

```python
img_f = img_as_float(img) # преобразование изображения в вещественное число
```

```python
from numpy import clip, array_equal
```

```python
imshow(clip(1 - img_f, 0 , 1));
```

```python
array_equal(clip(1 - img_f, 0 , 1), img_negative)
```
