```python
from skimage.io import imread, imsave, imshow
```

```python
img = imread('tiger-color.png')
```

```python
from skimage import img_as_float
```

```python
img_f = img_as_float(img)
```

```python
r = img_f[:, :, 0] # красный канал, двоеточие - все строки и все столбцы в изображении
```

```python
imshow(r);
```

```python
g = img_f[:, :, 1] # зелёный канал
```

```python
g.shape
```

```python
imshow(g);
```

```python
b = img_f[:, :, 2] # синий канал
```

```python
imshow(b);
```

```python
from numpy import dstack
```

```python
img_combined = dstack((r, g, b))
```

```python
imshow(img_combined);
```

```python
img_combined = dstack((g, r, b)) # поменяли цветовые каналы
```

```python
imshow(img_combined);
```

```python
img_combined = dstack((b, r, g)) # поменяли цветовые каналы
```

```python
imshow(img_combined);
```

```python
avg_gray = (r + g + b) / 3 # суммировали значения попиксельно, а затем все пиксели поделили на 3
```

```python
imshow(avg_gray); # усреднили 3 канала, получили серое изображение
```
