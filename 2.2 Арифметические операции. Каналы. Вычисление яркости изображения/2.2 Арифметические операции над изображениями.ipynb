```python
from skimage.io import imread, imsave, imshow
```

```python
img = imread('tiger-color.png')
```

```python
img.dtype # 8 бит - от 0 до 255
```

```python
from skimage import img_as_float
```

```python
img_f = img_as_float(img) # преобразование изображения в вещественное число
```

```python
img_f.dtype
```

```python
img.min(), img.max()
```

```python
img_f.min(), img_f.max()
```

```python
imshow(img_f / 1.5); # все цвета уменьшились - стали темнее
```

```python
imshow(img_f / 2);
```

```python
imshow(img_f / 4);
```

```python
imshow(img_f / 4 + 0.25);
```

```python
from numpy import clip # принимает нижний порог и верхний порог, значение ниже порога делает равными нижнему порогу
```

```python
imshow(clip(img_f * 1.5, 0 , 1));
```

```python
imshow(clip(img_f * 2, 0 , 1));
```

```python
imshow(clip(img_f * 2 - 0.5, 0, 1));
```
