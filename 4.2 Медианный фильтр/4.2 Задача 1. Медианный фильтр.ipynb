```python
from skimage.io import imread, imshow, imsave
```

```python
img = imread('tiger-gray-small.png')
```

```python
imshow(img);
```

```python
img.shape
```

```python
k = 7
```

```python
half = k // 2 
```

```python
half
```

```python
n_rows, n_columns = (img.shape[0], img.shape[1])
```

```python
out_img = np.zeros((n_rows, n_columns))
```

```python
for i in range(half, n_rows - half):
    for j in range(half, n_columns - half):
        box = img[i - half: i + half + 1, j - half: j + half + 1]
        median = np.median(box)
        out_img[i, j] = median
```

```python
out_img.min(), out_img.max()
```

```python
out_img.dtype
```

```python
out_img[3]
```

```python
out_img = out_img[half:n_rows - half, half:n_columns - half].astype('uint8')
```

```python
out_img
```

```python
img_median = imread('median-tiger.png')
```

```python
imshow(img_median);
```

```python
np.array_equal(out_img, img_median)
```

```python
imshow(out_img);
```

```python
imsave('tiger-median-my.png', out_img)
```
