```python
from skimage.io import imread, imshow, imsave
```

```python
img = imread('tiger-color.png')
```

```python
imshow(img);
```

```python
img.shape
```

```python
import scipy.fftpack as fp
```

```python
from skimage.color import rgb2gray
```

```python
from skimage import img_as_float
```

```python
freq_img = fp.fft2(rgb2gray(img))
```

```python
#freq_img = fp.fft2(img)
```

```python
freq_img.min(), freq_img.max(), freq_img.shape
```

```python
import numpy as np
```

```python
freq = np.log(1 + np.abs(fp.fftshift(freq_img)))
```

```python
freq.shape, freq.min(), freq.max(), freq.dtype
```

```python
imshow(freq)
```

```python
imsave('tiger-color-freq.png', freq)
```

```python
imshow(freq, cmap='gray')
```

```python
r, g, b = np.dsplit(img, 3)
```

```python
y = 0.2126 * r + 0.7152 * g + 0.0722 * b
```

```python
freq_img_y = fp.fft2(y)
```

```python
freq_y = np.log(1 + np.abs(fp.fftshift(freq_img_y)))
```

```python
imshow(freq_y)
```

```python
imshow(freq_y, cmap='gray')
```
