```python
from skimage.io import imread, imshow, imsave
```

```python
img = imread('apple.png')[..., :3]
```

```python
imshow(img);
```

```python
img.shape
```

```python
import numpy as np
```

```python
mask = np.zeros((img.shape[0], img.shape[1], img.shape[2]))
```

```python
mask.shape, mask.dtype
```

```python
mask
```

```python
imshow(mask);
```

```python
mask[:, :mask.shape[1] // 2, :] = 1.0
```

```python
imshow(mask);
```

```python
mask.dtype, mask.shape
```

```python
from skimage import img_as_ubyte
```

```python
mask = img_as_ubyte(mask)
```

```python
mask.dtype, mask.shape
```

```python
imshow(mask);
```

```python
imsave('mask-white-black.png', mask)
```

```python
img_one = img_as_ubyte(np.ones((img.shape[0], img.shape[1], img.shape[2])))
```

```python
imshow(img_one)
```
