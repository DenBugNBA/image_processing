```python
from skimage.io import imread, imshow, imsave
```

```python
img = imread('tiger-low-contrast.png')
```

```python
img.size
```

```python
img.min()
```

```python
imshow(img);
```

```python
img_f = img.astype('float')
```

```python
img_f.dtype
```

```python
k = round(img_f.size * 0.05)
```

```python
k
```

```python
pixels = img_f.ravel().copy()
```

```python
pixels.sort()
```

```python
pixels
```

```python
min_p, max_p = pixels[k + 1], pixels[-(k + 1)]
```

```python
min_p, max_p
```

```python
img_contrast = (img_f - min_p) * (255 / (max_p - min_p))
```

```python
img_contrast
```

```python
import numpy as np
```

```python
img_contrast = np.clip(img_contrast, 0, 255)
```

```python
img_final = img_contrast.astype('uint8')
```

```python
imshow(img_final)
```

```python
img_stable = imread('tiger-stable-contrast.png')
```

```python
np.array_equal(img_stable, img_final)
```
