```python
from skimage.io import imread, imshow
```

```python
img = imread('tiger-gray.png')
```

```python
imshow(img)
```

```python
from matplotlib.pyplot import hist
```

Функция [`matplotlib.pyplot.hist`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html) используется для построения гистограммы массива чисел. Для использования функции с изображениями массив нужно вытянуть в одномерный с помощью метода [`ravel`](https://docs.scipy.org/doc/numpy/reference/generated/numpy.ravel.html). С помощью аргумента `bins` задается количество или границы корзин гистограммы.

Функция построения гистограммы возвращает кортеж из трех элементов — значений гистограммы, границы корзин и отдельные прямоугольники, использовавшиеся для построения гистограммы.



```python
values, bin_edges, patches = hist(img.ravel(), bins=range(257))
```

```python
values
```

```python
bin_edges
```

```python
patches
```

Поведение функции `matplotlib.pyplot.hist` аналогично функции `numpy.histogram`. Если вам не нужно строить график, а нужно только посчитать гистограмму, используйте эту функцию.

```python
from numpy import histogram
values, bin_edges = histogram(img, bins=range(257))
```
