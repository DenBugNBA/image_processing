```python
from skimage.io import imread, imshow, imsave
```

```python
img = imread('tiger-low-contrast.png')
```

```python
imshow(img);
```

```python
from matplotlib.pyplot import hist
```

```python
values, bin_edges, patches = hist(img.ravel(), bins=range(257))
```

```python
img.min()
```

```python
img.max()
```

```python
img_contrast = (img - img.min()) * (255 / (img.max() - img.min()))
```

```python
values, bin_edges, patches = hist(img_contrast.ravel(), bins=range(257))
```

```python
img_contrast.dtype
```

```python
img_contrast = img_contrast.astype('uint8')
```

```python
img_contrast.dtype
```

```python
imshow(img_contrast);
```

```python
img_high_contrast = imread('tiger-high-contrast.png')
```

```python
import numpy as np
```

```python
np.array_equal(img_contrast, img_high_contrast)
```

```python
imsave('tiger-high-contrast-my.png', img_contrast)
```
