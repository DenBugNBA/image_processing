```python
from skimage.io import imread, imshow, imsave
```

```python
img = imread('tiger-gray-small.png', 1)
```

```python
# img = imread('sample.jpg', 1)
```

```python
img.dtype
```

```python
imshow(img);
```

```python
img.shape
```

```python
import numpy as np
```

```python
kernel_size = 5
```

```python
kernel = np.ones((kernel_size, kernel_size))
```

```python
kernel
```

```python
from scipy.signal import convolve2d
```

```python
img_filtered = (convolve2d(img, kernel, mode='valid') / kernel_size ** 2).astype('uint8')
```

```python
img_filtered.shape
```

```python
img_filtered.dtype
```

```python
imshow(img_filtered);
```

```python
img_filtered.min(), img_filtered.max()
```

```python
img_box = imread('box-tiger.png')
```

```python
imshow(img_box);
```

```python
import numpy as np
```

```python
np.array_equal(img_filtered, img_box)
```

```python
img_box.min(), img_box.max()
```
