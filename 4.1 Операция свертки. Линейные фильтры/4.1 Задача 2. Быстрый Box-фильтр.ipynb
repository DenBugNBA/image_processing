```python
from skimage.io import imread, imshow, imsave
```

```python
img = imread('tiger-gray-small.png')
```

```python
imshow(img);
```

```python
img.shape
```

```python
from skimage.transform import integral_image, integrate
```

```python
img_integral = integral_image(img)
```

```python
img_integral.min(), img_integral.max()
```

```python
# imshow(img_integral, cmap='gray');
```

```python
img_integral_normalized = (img_integral - img_integral.min()) / (img_integral.max() - img_integral.min())
```

```python
img_integral_normalized.min(), img_integral_normalized.max()
```

```python
imshow(img_integral_normalized);
```

```python
from skimage import img_as_ubyte
```

```python
img_integral_normalized_byte = img_as_ubyte(img_integral_normalized)
```

```python
img_integral_normalized_byte.min(), img_integral_normalized_byte.max()
```

```python
imshow(img_integral_normalized_byte);
```
