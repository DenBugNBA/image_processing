```python
from skimage.io import imread, imshow, imsave
```

```python
img = imread('sample.jpg')
```

```python
imshow(img);
```

```python
from skimage import img_as_float, img_as_ubyte
```

```python
img_f = img_as_float(img)
```

```python
img_f.dtype
```

```python
img_f
```

```python
import numpy as np
```

```python
r, g, b = np.dsplit(img_f, 3)
```

```python
r.shape
```

```python
r.size
```

```python
r.ndim
```

```python
r_avg, g_avg, b_avg = [np.mean(i) for i in (r, g, b)]
```

```python
r_avg
```

```python
g_avg
```

```python
b_avg
```

```python
avg = np.mean((r_avg, g_avg, b_avg))
```

```python
avg
```

```python
r_w, g_w, b_w = [i / avg for i in (r_avg, g_avg, b_avg)]
```

```python
r_new = r / r_w
```

```python
g_new = g / g_w
```

```python
b_new = b / b_w
```

```python
r_new.min(), r_new.max()
```

```python
img_grey_world = np.dstack((r_new, g_new, b_new))
```

```python
img_grey_world = np.clip(img_grey_world, 0, 1)
```

```python
imshow(img_grey_world);
```

```python
img_grey_world_final = img_as_ubyte(img_grey_world)
```

```python
imsave('grey-world-my.jpg', img_grey_world_final)
```
